# Tópico 1 — Criação de arrays NumPy

Neste tópico, vamos começar a trabalhar com `NumPy`, uma das bibliotecas mais importantes para Ciência de Dados em Python.

Comparação com C:

- uma `list` em Python se comporta como uma sequência de **referências para objetos**;
- um `np.ndarray` armazena dados homogêneos em uma estrutura muito mais próxima de um **vetor tipado e contíguo em memória**, semelhante ao que fazemos em C com arrays de `int`, `float` ou `double`;
- em C, o programador gerencia diretamente a memória com variáveis, ponteiros, `malloc` e `free`;
- em NumPy, o objeto `ndarray` encapsula esse bloco de memória e guarda metadados como tipo dos dados, dimensão e formato.

Antes de fazer operações matemáticas avançadas com arrays, precisamos saber criá-los corretamente.

Neste tópico, vamos usar principalmente:


# Exercício 5 — Organizando os dados em uma matriz do cardápio

Agora que já criamos arrays separados para códigos, preços, quantidades vendidas e disponibilidade, vamos combinar essas informações em uma única estrutura tabular.

A ideia é montar uma matriz NumPy em que:

- cada linha representa um produto;
- cada coluna representa uma informação sobre o produto.

A matriz seguirá esta ordem de colunas:

```python
[código, preço, quantidade vendida, disponibilidade]
```

Esse formato lembra uma matriz em C, como `double dados[5][4]`. Porém, existe um detalhe importante: arrays NumPy comuns possuem dados homogêneos. Como uma das colunas possui valores `float64`, o NumPy converte a matriz final para um tipo numérico comum.

### Requisitos

- Crie um array `codigos_produtos` com os códigos de `100` a `104`.
- Crie um array `precos` com 5 preços.
- Crie um array `quantidades_vendidas` preenchido inicialmente com zeros.
- Crie um array `produtos_disponiveis` preenchido inicialmente com uns.
- Combine esses arrays em uma matriz chamada `dados_cardapio`.
- A matriz deve ter 5 linhas e 4 colunas.
- Exiba:
  - a matriz completa;
  - o formato da matriz;
  - o tipo dos dados armazenados na matriz.
- Exiba os dados linha por linha, associando cada linha ao nome do produto correspondente.

````python
import numpy as np
import numpy.typing as npt

nomes_produtos: list[str] = [
    "Café",
    "Pão de queijo",
    "Suco",
    "Sanduíche",
    "Bolo"
]

# Crie os arrays abaixo (já criados nos exercícios anteriores).
codigos_produtos: npt.NDArray[np.int64] = np.arange(100, 105, dtype=np.int64)

precos: npt.NDArray[np.float64] = np.array([4.50, 6.00, 7.50, 12.00, 5.50], dtype=np.float64)
quantidades_vendidas: npt.NDArray[np.int64] = np.zeros(5, dtype=np.int64)
produtos_disponiveis: npt.NDArray[np.bool_] = np.ones(5, dtype=np.int64)

# Combine os arrays como colunas de uma matriz.
dados_cardapio = np.column_stack((codigos_produtos, precos, quantidades_vendidas, produtos_disponiveis))

print("=== Matriz do cardápio ===")
print(dados_cardapio)

print("\n=== Estrutura da matriz ===")
print(f"Formato da matriz: {dados_cardapio.shape}")
print(f"Tipo dos dados armazenados: {dados_cardapio.dtype}")


print("\n=== Dados por produto ===")

indice_linha: int
linha: npt.NDArray[np.float64]

for indice_linha, linha in enumerate(dados_cardapio):
    nome: str = nomes_produtos[indice_linha]
    codigo: int = linha[0]
    preco: float = linha[1]
    vendidos: int = linha[2]
    disponivel: bool = linha[3]

    print(
        f"{nome} "
        f"código={codigo}, "
        f"preço=R$ {preco:.2f}, "
        f"vendidos={vendidos}, "
        f"disponível={disponivel}"
    )
````

O P.U. registrou a quantidade vendida de três produtos durante quatro dias.

# Exercício 3 — Transformando uma sequência em matriz

O P.U. recebeu os dados de vendas em uma sequência única, como se fosse um vetor linear.

Porém, sabemos que esses dados representam 4 dias e 3 produtos. Portanto, queremos reorganizar esse vetor em uma matriz 4x3.

Em C, isso exigiria cuidado com a forma como os dados estão armazenados linearmente na memória. Em NumPy, podemos usar `reshape()` para mudar a interpretação dimensional dos dados.

### Requisitos

- Crie um array chamado `vendas_lineares` com 12 valores.
- Use `np.arange()` para gerar valores de `1` até `12`.
- Reorganize esse array em uma matriz chamada `vendas_matriz`.
- A matriz deve ter 4 linhas e 3 colunas.
- Exiba o array original e a matriz reorganizada.
- Exiba o `shape` de ambos.

````python
import numpy as np
import numpy.typing as npt

vendas_lineares: npt.NDArray[np.int64] = np.arange(1, 13)

vendas_matriz: npt.NDArray[np.int64] = vendas_lineares.reshape((4, 3))

print(f"Vendas em formato linear: {vendas_lineares}")
print(f"shape do array linear: {vendas_lineares.shape}")

print("\nVendas em formato de matriz:")
print(vendas_matriz)
print(f"shape da matriz: {vendas_matriz.shape}")
````

# Exercício 4 — Interpretando dados por turno

O P.U. registrou vendas de dois turnos, manhã e tarde, para três produtos durante dois dias.

Os dados chegaram em formato linear, mas queremos reorganizá-los em uma estrutura com três dimensões:

- dimensão 0: dia;
- dimensão 1: turno;
- dimensão 2: produto.

Essa estrutura já começa a mostrar que arrays NumPy podem representar mais do que matrizes 2D.

### Requisitos

- Crie um array chamado `vendas_lineares` com 12 valores.
- Reorganize os dados em um array chamado `vendas_por_turno`.
- O formato final deve ser `(2, 2, 3)`.
- Exiba o array resultante.
- Exiba `ndim`, `shape` e `size`.

````python
import numpy as np
import numpy.typing as npt

# Dados lineares:
# 12, 8, 5, 10, 6, 4, 14, 9, 7, 11, 8, 6
vendas_lineares: npt.NDArray[np.int64] = np.array([12, 8, 5, 10, 6, 4, 14, 9, 7, 11, 8, 6], dtype=np.int64)
print(f"Vendas lineares: {vendas_lineares}")
# Reorganize os dados no formato (2, 2, 3).
vendas_por_turno: npt.NDArray[np.int64] = vendas_lineares.reshape((2, 2, 3))

print("Vendas por dia, turno e produto:")
print(vendas_por_turno)

print(f"\nndim: {vendas_por_turno.ndim}")
print(f"shape: {vendas_por_turno.shape}")
print(f"size: {vendas_por_turno.size}")
````

# Tópico 4 — Copy e view

Em NumPy, nem todo recorte cria uma cópia independente dos dados.

Quando usamos slicing, muitas vezes o resultado é uma **view**, ou seja, uma nova forma de enxergar o mesmo bloco de memória.

Isso é muito importante para quem vem de C.

Em C, quando passamos um ponteiro para uma função ou acessamos uma parte de um vetor por deslocamento de endereço, podemos alterar os dados originais sem criar uma cópia. Em NumPy, uma view tem uma ideia parecida: ela aponta para os mesmos dados internos do array original, embora o acesso seja mediado pelo objeto `ndarray`.

Neste tópico, vamos usar principalmente:

```python
array[inicio:fim]
array.copy()
array.base
np.shares_memory()
```

# Exercício 3 — Verificando compartilhamento de memória

O P.U. quer confirmar se uma fatia compartilha memória com o array original.

Para isso, podemos usar `np.shares_memory()`.

Esse comando é útil para entender se estamos lidando com uma view ou com uma cópia.

### Requisitos

- Crie um array `precos`.
- Crie uma fatia com slicing.
- Crie uma cópia com `.copy()`.
- Use `np.shares_memory()` para verificar:
  - se a fatia compartilha memória com o original;
  - se a cópia compartilha memória com o original.
- Exiba os resultados.

````python
import numpy as np
import numpy.typing as npt

# Lista de preços: 4.50, 6.00, 7.50, 12.00, 5.50
precos: npt.NDArray[np.float64] = np.array([4.50, 6.00, 7.50, 12.00, 5.50], dtype=np.float64)

# Crie uma fatia com slicing.
precos_fatia: npt.NDArray[np.float64] = precos[1:3]

# Crie uma cópia da mesma fatia.
precos_copia: npt.NDArray[np.float64] = precos_fatia.copy()

# Verifique o compartilhamento de memória.
fatia_compartilha_memoria: bool = np.shares_memory(precos_fatia, precos)
copia_compartilha_memoria: bool = np.shares_memory(precos_copia, precos)

print(f"A fatia compartilha memória com o original? {fatia_compartilha_memoria}")
print(f"A cópia compartilha memória com o original? {copia_compartilha_memoria}")
````

# Tópico 5 — Fancy indexing e máscaras booleanas

Além da indexação comum e do slicing, NumPy permite selecionar dados usando listas ou arrays de índices.

Isso é chamado de **fancy indexing**.

Também podemos selecionar dados usando condições, criando **máscaras booleanas**.

Esses dois recursos são muito importantes em Ciência de Dados, porque permitem filtrar dados sem escrever laços explícitos.

Em C, normalmente precisaríamos percorrer o vetor com `for`, testar condições com `if` e copiar manualmente os elementos desejados para outro vetor. Em NumPy, fazemos isso de forma mais direta.

Neste tópico, vamos usar principalmente:

```python
array[[indices]]
matriz[[linhas]]
array[mascara]
array[condicao]
```

````python
import numpy as np
import numpy.typing as npt

# Lista de preços: 4.50, 6.00, 7.50, 12.00, 5.50
precos: npt.NDArray[np.float64] = np.array([4.50, 6.00, 7.50, 12.00, 5.50], dtype=np.float64)

# Índices escolhidos: 1, 2 e 3
indices: npt.NDArray[np.int64] = np.array([1, 2, 3], dtype=np.int64)

# Selecione os preços usando slicing.
precos_slicing: npt.NDArray[np.float64] = precos[1:4]

# Selecione os preços usando fancy indexing.
precos_fancy: npt.NDArray[np.float64] = precos[indices]

# Verifique compartilhamento de memória.
slicing_compartilha_memoria: bool = np.shares_memory(precos, precos_slicing)
fancy_compartilha_memoria: bool = np.shares_memory(precos, precos_fancy)

print(f"Resultado com slicing: {precos_slicing}")
print(f"Resultado com fancy indexing: {precos_fancy}")

print(f"\nSlicing compartilha memória com o original? {slicing_compartilha_memoria}")
print(f"Fancy indexing compartilha memória com o original? {fancy_compartilha_memoria}")
````
